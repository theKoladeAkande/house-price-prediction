```python
# seperate data types
def seperate_types(data :"A pandas DataFrame" ):
    """Seperate data into numeric and qualitative based on data type"""
    numeric = []
    qualitative = []
    for i in data.columns:
        if data[i].dtype in [np.int32, np.int64, np.float32, np.float64] and (i != 'Id'):
            numeric.append(i)
        elif data[i].dtype == object:
            qualitative.append(i)
    return numeric , qualitative
```

## Missing Values

```python
# check and calculate % of missing value 
def calculate_missing(data: "a pandas data frame"):
    """ calculates and returns  a pandas DataFrame of % and number of missing value"""
    missing_data_info ={}
    length_ = data.shape[0]
    for i in data.columns:
        missing_series = data[i].isnull().value_counts()
        if missing_series.shape[0] == 2:
            missing_count = data[i].isnull().value_counts()[True]
            percentage = (missing_count/length_) * 100
            missing_data_info[i] = [missing_count, percentage]
        else:
            missing_data_info[i]=[0,0]
        
    return pd.DataFrame(missing_data_info, index=['missing_count','%']).T
       
```

```python
def check_cardinality(data, qualitative_list ):
    """computes the cardinality of the categorical variables """
    cardinality_count = { }
    for i in qualitative_list:
        cardinal_no=data[i].value_counts().shape[0]
        cardinality_count[i] = cardinal_no
    return pd.Series(cardinality_count)
    
```

```python
train_data_cardinality =check_cardinality(train_data, qualitative)
test_data_cardinality =check_cardinality(test_data, qualitative)
display(train_data_cardinality[train_data_cardinality >= 7])
print("-----------Test")
display(test_data_cardinality[test_data_cardinality >= 7])
```

```python
from collections import defaultdict
def rare_values(data, cat_vars, value):
    rare_list = defaultdict(list)
    for i in cat_vars:
        data_rare = data[i].value_counts()/len(data)
        rare_list[i].append(data_rare[data_rare< value].index)
        rare_list[i].append(data_rare[data_rare< value].values)
    return rare_list
```
