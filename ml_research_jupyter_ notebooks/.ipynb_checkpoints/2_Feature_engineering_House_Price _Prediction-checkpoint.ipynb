```python
#functions 

def seperate_types(data :"A pandas DataFrame" ):
    """Seperate data into numeric and qualitative based on data type"""
    numeric = []
    qualitative = []
    for i in data.columns:
        if data[i].dtype in [np.int32, np.int64, np.float32, np.float64] and (i != 'Id'):
            numeric.append(i)
        elif data[i].dtype == object:
            qualitative.append(i)
    return numeric , qualitative


def calculate_missing(data: "a pandas data frame"):
    """ calculates and returns  a pandas DataFrame of % and number of missing value"""
    missing_data_info ={}
    length_ = data.shape[0]
    for i in data.columns:
        missing_series = data[i].isnull().value_counts()
        if missing_series.shape[0] == 2:
            missing_count = data[i].isnull().value_counts()[True]
            percentage = (missing_count/length_) * 100
            missing_data_info[i] = [missing_count, percentage]
        else:
            missing_data_info[i]=[0,0]
        
    return pd.DataFrame(missing_data_info, index=['missing_count','%']).T
```

```python
numeric , qualitative = seperate_types(train_X)
```

```python
qualitative_missing_df=calculate_missing(train_X[qualitative])
#get the categorical features with greater than 0 but less than 80 missing values
qualitative_missing_gt0_lt80 = qualitative_missing_df[(qualitative_missing_df['%']>0) & (qualitative_missing_df['%']< 80) ]
```

```python
#get the categorical features with greater than 80 missing values
qualitative_missing_gt80 = qualitative_missing_df[ qualitative_missing_df['%'] >80 ]
```

# Categorical 

```python
def fill_categorical_na(data, cat_missing_cols):
    """ Creates another category in categorical features by filling missing values with 'MissingValue' """
    df_copy = data.copy()
    df_copy[cat_missing_cols] = data[cat_missing_cols].fillna("MissingValue") 
    return df_copy
```

```python
train_X = fill_categorical_na(train_X, qualitative_missing_gt0_lt80.index)
test_X = fill_categorical_na(test_X, qualitative_missing_gt0_lt80.index)
```

```python
train_X[qualitative_missing_gt0_lt80.index].isnull().sum()
```

```python
test_X[qualitative_missing_gt0_lt80.index].isnull().sum()
```

```python
train_X2 = pd.concat([train_X,train_y], axis=1)
```

**for observations in the categorical features of our datasets that have very low appearance < 1%, they are handled seperately**

```python
def rare_values(data,cat_col,value):
    """ Returns the categories for each feature greater with than 1% observation in the feature"""
    data_copy = data.copy()
    df = data_copy[[cat_col,'SalePrice']].groupby(cat_col).count()/len(data_copy)
    return df[df["SalePrice"]>value].index
```

```python
for i in qualitative:
    #note that this is done on the train set 
    rare_data = rare_values(train_X2,i, 0.01)
    train_X[i] = np.where(train_X[i].isin(rare_data), train_X[i], 'Rare')
    test_X[i] = np.where(test_X[i].isin(rare_data), test_X[i], 'Rare')
```

**In this regression task, to help capture the relationship between the categorical features and the target variables a monotonic ordering is applied on the categorical features. The Features are grouped by the median of the target(to account for outliers) and are sorted based on the median in ascending order, the observation in the sorted features are labeled from 0-n(n-the cardinality of the feature).
This is thought to be a better transformation than one-hot-encoding,most  especially  because tree models are considered for this project**

```python
train_X = pd.concat([train_X, train_y], axis =1)
def monothonic_ordering(train_set, test_set, cat_col, target):
    """compute a monothonic ordering for categorical features"""
    monothonic_order = train_set[[target, cat_col]].groupby(cat_col).median().sort_values(target).index
    ordered_label = {j:i for i,j in enumerate(monothonic_order,start=0)}
    train_set[cat_col] = train_set[cat_col].map(ordered_label)
    test_set[cat_col] = test_set[cat_col].map(ordered_label)
```

```python
for i in qualitative:
    monothonic_ordering(train_X, test_X, i,"SalePrice")
```

```python
def year_difference(data,yr_col):
    """Capture the diffrence in between Feature Year sold and Year column"""
    data[yr_col] = data['YrSold'] - data[yr_col]
    return data
```

# Creating other Features 

```python
def binary_area_feature(data, col):
    data['has'+col] = data[col].apply(lambda x: 1 if x > 0 else 0)
    return data
```
